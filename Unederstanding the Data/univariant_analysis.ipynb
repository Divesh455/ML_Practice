```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.sample(5)
```

# Categorical Data

## a. Countplot

```python
sns.countplot(data=df, x='Pclass')
```

## b.PieChart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# Numerical Data

## a.Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'],bins=6)
```

## b. Distplot

```python
sns.distplot(df['Age'])
```

## c.Boxplot

```python
sns.boxplot(df['Age'])
```

```python
df['Age'].mean()
```

```python
df['Age'].skew()
```
