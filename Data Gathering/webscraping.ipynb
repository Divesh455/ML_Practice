```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup = BeautifulSoup(webpage)
```

```python
# print(soup.prettify())
```

```python
# soup.find_all('h1')[0].text
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
company_name = []
company_rating = []
company_review = []
avg_sal = []
services = []
locations = []
for i in company:
    name = i.find('h2', class_='companyCardWrapper__companyName').text.strip()
    company_name.append(name)

    rating = i.find("div", class_="rating_text rating_text--md")
    company_rating.append(rating.get_text(strip=True))

    review = i.find_all('span', class_='companyCardWrapper__ActionCount')[0]
    company_review.append(review.text)

    sal = i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text
    avg_sal.append(sal)

    sevices = i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0]
    services.append(sevices)

    location = i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[1].split('+')[0].split(' /')[0]
    locations.append(location)
```

```python
print(company_name)
print(company_rating)
print(company_review)
print(avg_sal)
print(services)
print(locations)
```

```python
company_name = []
company_rating = []
company_review = []
avg_sal = []
services = []
locations = []
about_company = []
for i in range(1,16):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(f'https://www.ambitionbox.com/list-of-companies?page={i}',headers=headers).text

    soup = BeautifulSoup(webpage)
    company = soup.find_all('div', class_='companyCardWrapper')

    for j in company:
        name = j.find('h2', class_='companyCardWrapper__companyName').text.strip()
        company_name.append(name)
    
        rating = j.find("div", class_="rating_text rating_text--md")
        company_rating.append(rating.get_text(strip=True))
    
        review = j.find_all('span', class_='companyCardWrapper__ActionCount')[0]
        company_review.append(review.text)
    
        sal = j.find_all('span',class_='companyCardWrapper__ActionCount')[1].text
        avg_sal.append(sal)
    
        sevices = j.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0]
        services.append(sevices)
    
        location = j.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[1].split('+')[0].split(' /')[0]
        locations.append(location)

    lower_name = []
    for k in company_name:
        lower_name.append(k.lower())

    for l in lower_name:
        about_webpage = requests.get(f'https://www.ambitionbox.com/overview/{l}-overview',headers=headers).text
        new_soup = BeautifulSoup(about_webpage)
        about = new_soup.find('div',class_='text-sm font-pn-400 [&_ul]:list-disc [&_ol]:list-[auto] [&_ul]:ml-5 [&_ol]:ml-5') 
        about_company.append(about.text.split('.')) if about != None else about_company.append('None')

d = { 'company_name':company_name,'company_rating':company_rating,'company_review':company_review,'avg_sal':avg_sal,'services':services,'location':locations,'about_company':about_company}
df = pd.DataFrame(d)
```
