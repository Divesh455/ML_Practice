```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# Step 1: Load data
df = pd.read_csv('cars.csv')

# Step 2: Select feature (you can add more later)
X = df[['brand']]

# Step 3: Train-test split
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Step 4: Handle rare categories (ONLY using training data)
threshold = 100
counts = X_train['brand'].value_counts()
rare_brands = counts[counts <= threshold].index

# Replace rare with 'uncommon'
X_train['brand'] = X_train['brand'].replace(rare_brands, 'uncommon')
X_test['brand'] = X_test['brand'].replace(rare_brands, 'uncommon')

# Step 5: OneHot Encoding
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

X_train_encoded = ohe.fit_transform(X_train)
X_test_encoded = ohe.transform(X_test)

# Step 6: Convert to DataFrame (clean output)
X_train_final = pd.DataFrame(
    X_train_encoded,
    columns=ohe.get_feature_names_out(['brand'])
)

X_test_final = pd.DataFrame(
    X_test_encoded,
    columns=ohe.get_feature_names_out(['brand'])
)

# Reset index (important)
X_train_final.reset_index(drop=True, inplace=True)
X_test_final.reset_index(drop=True, inplace=True)

# Final datasets ready ✅
X_train_final.head()
# print(X_test_final.head())
```
